```python
# Write a program that finds all elements appearing in both lists (duplicates allowed only once).
# Example:
# List1 = [1, 2, 3, 4]
# List2 = [3, 4, 5, 6]
# Output â†’ [3, 4]

list1 = [1, 2, 3, 4]
list2 = [3, 4, 5, 6]
common=[]

for i in list1:
    if i in list2 and i not in common:
        common.append(i)

print(common)
```

```python
# Given a sentence, write a program that counts how many times each word appears.
# Example:
# Input: "the cat and the hat"
# Output: {'the': 2, 'cat': 1, 'and': 1, 'hat': 1}

s="the cat and the hat"
words=s.split()
freq={}
for w in words:
    if w in freq:
        freq[w]=freq[w]+1
    else:
        freq[w]=1
print(freq)
```

```python
# Given a 2D list (matrix), write Python code that prints its transpose.
# Example:
# Input: [[1, 2], [3, 4], [5, 6]]
# Output: [[1, 3, 5], [2, 4, 6]]

matrix=[[1,2],[3,4],[5,6]]
transpose=[]
for col in range(len(matrix[0])):
    new_row=[]
    for row in range(len(matrix)):
        new_row.append(matrix[row][col])
    transpose.append(new_row)
print(transpose)
```

```python
# Write a program that reverses the order of words in a string.
# Example:
# Input: "hello world python"
# Output: "python world hello"
s="hello world python"
words=s.split()
rev=" ".join(words[::-1])
print(rev)
```

```python
# Write a Python program that removes duplicates from a list without using set(),
# while preserving the original order.
# Example:
# Input: numbers = [4, 2, 4, 5, 2, 3, 5, 1]
# Output: [4, 2, 5, 3, 1]
nums=[4,2,4,5,2,3,5,1]
new=[]
for i in nums:
    if i not in new:
        new.append(i)
print(new)
```

```python
# Given a list of numbers and a target sum, find all unique pairs that add up to that sum.
# Example:
# Input:
# List -> [2, 4, 3, 5, 7, 8, -1]
# Target sum -> 6
# Output: [(2, 4), (3, 3), (7, -1)]

nums = [2,4,3,5,7,8,-1] 
target = 6 
pairs = [] 
for i in range(len(nums)): 
    for j in range(i,len(nums)): 
        if nums[i] + nums[j] == target: 
            pair=(nums[i], nums[j])
            sorted_pair=tuple(sorted(pair))
            if sorted_pair not in pairs:
                pairs.append(sorted_pair)

print(pairs)
```
